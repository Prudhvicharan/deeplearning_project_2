## CNN Implementation

```python
import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from keras.optimizers import SGD, Adam, RMSprop
from keras.utils import to_categorical
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt

def load_data():
    (x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
    x_train = x_train / 255.0
    x_test = x_test / 255.0
    y_train = to_categorical(y_train)
    y_test = to_categorical(y_test)
    return (x_train, y_train), (x_test, y_test)

def create_cnn_model(config):
    model = Sequential()
    model.add(Conv2D(config['filters'], config['kernel_size'], activation='relu', input_shape=(28, 28, 1), padding='same'))
    if config['dropout']:
        model.add(Dropout(config['dropout_rate']))
    model.add(MaxPooling2D(config['pool_size']))

    for _ in range(config['num_layers'] - 1):
        model.add(Conv2D(config['filters'], config['kernel_size'], activation='relu', padding='same'))
        if config['dropout']:
            model.add(Dropout(config['dropout_rate']))
        model.add(MaxPooling2D(config['pool_size']))

    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    if config['dropout']:
        model.add(Dropout(config['dropout_rate']))
    model.add(Dense(10, activation='softmax'))

    return model

def compile_model(model, optimizer, learning_rate):
    if optimizer == SGD:
        opt = SGD(learning_rate=learning_rate)
    elif optimizer == Adam:
        opt = Adam(learning_rate=learning_rate)
    elif optimizer == RMSprop:
        opt = RMSprop(learning_rate=learning_rate)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

def train_model(model, x_train, y_train, epochs, batch_size):
    history = model.fit(x_train.reshape(-1, 28, 28, 1), y_train, epochs=epochs, batch_size=batch_size, verbose=1, validation_split=0.2)  # Add validation split
    return history

def evaluate_model(model, x_test, y_test):
    return model.evaluate(x_test.reshape(-1, 28, 28, 1), y_test, verbose=2)

# Load data
(x_train, y_train), (x_test, y_test) = load_data()

# Define configurations
configurations = [
    {'num_layers': 2, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': False},
    {'num_layers': 2, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': SGD, 'learning_rate': 0.01, 'dropout': False},
    {'num_layers': 2, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': RMSprop, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 2, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 3, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': SGD, 'learning_rate': 0.01, 'dropout': True, 'dropout_rate': 0.3},
    {'num_layers': 3, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': False},
    {'num_layers': 3, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': RMSprop, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.3},
    {'num_layers': 2, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': SGD, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 2, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': False},
    {'num_layers': 2, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': RMSprop, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 2, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': SGD, 'learning_rate': 0.01, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 3, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': False},
    {'num_layers': 3, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': RMSprop, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.2},
    {'num_layers': 3, 'filters': 64, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 64, 'optimizer': SGD, 'learning_rate': 0.001, 'dropout': True, 'dropout_rate': 0.3},
    {'num_layers': 2, 'filters': 32, 'kernel_size': (3, 3), 'pool_size': (2, 2), 'batch_size': 32, 'optimizer': Adam, 'learning_rate': 0.001, 'dropout': False},
]


CNN_results = []
best_config = None
best_accuracy = 0

for i, config in enumerate(configurations, start=1):
    print(f"Configuration-{i}")
    print(f"Data: Layers=[{config['num_layers']}], Filters=[{config['filters']}], Kernel Size={config['kernel_size']}, Pool Size={config['pool_size']}, Batch Size={config['batch_size']}, Optimizer={config['optimizer'].__name__}, Learning Rate={config['learning_rate']}, Dropout={config['dropout']}, Dropout Rate={config['dropout_rate'] if config['dropout'] else 'N/A'}")

    # Create CNN model
    model = create_cnn_model(config)

    # Compile model
    compile_model(model, config['optimizer'], config['learning_rate'])

    # Train model
    print("Training...")
    history = train_model(model, x_train, y_train, epochs=5, batch_size=config['batch_size'])

    # Evaluate model
    test_loss, test_acc = evaluate_model(model, x_test, y_test)
    print(f"Finished. Accuracy: {test_acc:.4f}")
    print("------------------------------------------")

    CNN_results.append((config, test_acc))

    if test_acc > best_accuracy:
        best_accuracy = test_acc
        best_config = config

    # Plot training and validation accuracy curves
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # Plot training and validation loss curves
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

# Tabulate results
print("\nResults:")
headers = ["Configuration", "Accuracy"]
table = tabulate([(idx + 1, acc) for idx, (cfg, acc) in enumerate(CNN_results)], headers=headers, tablefmt="grid")
print(table)

# Display best configuration and visualize
print(f"\nBest Configuration: {best_config}")
print(f"Best Accuracy: {best_accuracy:.4f}")

# Visualize the best model's performance (you can add your visualization code here)
# For example, you can plot the learning curves or display sample predictions
```

## Plotting All Configurations

```python
# Extract configurations and accuracies from CNN_array
configurations = [config[0] for config in CNN_results]
accuracies = [result[1] * 100 for result in CNN_results]

# Plotting accuracy vs configurations
plt.figure(figsize=(10, 6))
bars = plt.bar(range(1, len(configurations) + 1), accuracies, color='skyblue')
plt.xlabel('Configuration')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy vs Configurations')
plt.xticks(range(1, len(configurations) + 1), range(1, len(configurations) + 1), rotation=45, ha='right')
plt.ylim(0, 100)  # Setting y-axis limit from 0 to 100 for percentage
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with accuracy percentages
for bar, acc in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 5, f'{acc:.2f}%', ha='center', va='bottom', color='black', fontsize=8)

plt.tight_layout()
plt.show()
```

## Plotting the Best Configuration

```python
# Find the configuration with the highest accuracy
best_config_num = CNN_results.index(max(CNN_results, key=lambda x: x[1])) + 1
best_config = max(CNN_results, key=lambda x: x[1])[0]
best_accuracy = max(CNN_results, key=lambda x: x[1])[1]

# Convert accuracy to percentage without decimal places
accuracy_percent = best_accuracy * 100

# Select hyperparameters to plot
selected_hyperparameters = ['num_layers', 'filters', 'batch_size', 'learning_rate']
selected_values = [best_config[param] for param in selected_hyperparameters]

# Plot each hyperparameter separately
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle(f"Configuration {best_config_num} (Accuracy: {accuracy_percent:.2f}%)", fontsize=16)

for i, param in enumerate(selected_hyperparameters):
    ax = axs[i//2][i%2]
    ax.bar(param, selected_values[i], color='skyblue')
    ax.set_title(param.capitalize())
    ax.set_ylabel('Values')

plt.tight_layout()
plt.show()
```
